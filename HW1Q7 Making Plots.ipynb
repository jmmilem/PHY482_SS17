```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import ma
%matplotlib inline
plt.style.use('fivethirtyeight')

x = np.arange(-5,5,0.3)
y = np.arange(-5,5,0.3)
X, Y = np.meshgrid(x,y)
e = 1.602*10**(-19)
v = e/(2)*np.sqrt((x**2+y**2))**2

plt.figure(figsize=(8,8))
plt.quiver(X,Y,v, scale=100)
plt.xlim([-50,50])
plt.ylim([-50,50])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import ma
%matplotlib inline
plt.style.use('fivethirtyeight')

x = np.linspace(-0.5,0.5,50)
y = np.linspace(-0.5,0.5,50)
X, Y = np.meshgrid(x,y)

e = 1.602*10**(-19)
r = np.sqrt(X**2+Y**2)
alpha = 1/(3.14*4*e)

vx = (3*X*Y**2/r**5)*alpha
vy = alpha*(3*Y**2/r**5-1/r**3)

plt.figure(figsize=(8,8))
plt.quiver(X,Y,vx,vy)
plt.ylim([-0.2,0.2])

plt.show()
```
